menghapus kolom yang valuenya kosong semua

```python
wea_1=wea.drop(['Formatted Date','Loud Cover'], axis=1)
wea_1.head(10)
```

```python
wea_1.info()
```

visualisasi untuk mencari mana yang berpengaruh karena kolerasinya

```python
plt.figure(figsize=(10,10))  # to get size of your plot, you can change with any numbers
sns.heatmap(wea_1.corr(), # data to be shown in the plot
            annot=True, # do you want to get value instead of just color?
            linewidths=.10, # line between each box
            fmt= '.2f', # how many 0 behind coma (f for decimal)
            cmap='Blues'); # color map
```

suhu dipengaruhi oleh visibility

SINGLE LINEAR REGRESSION

memecah jadi training dan testing

```python
# memanggil package linear regresi
from sklearn.linear_model import LinearRegression
# memanggil package train test split
from sklearn.model_selection import train_test_split

# data target (X)
X = wea_1[['Visibility (km)']]
# data fitur (y)
y = wea_1['Temperature (C)']

# pembagian data dengan train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, random_state=45)

# cari tau shapenya
# data training
print("Training Data :", X_train.shape)
# data testing
print("Testing Data :", X_test.shape)
```

menfit data model trainingnya

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

cetak score dari evaluasi data trainingnya

```python
print("Evaluasi dari Data Training :",lr.score(X_train, y_train))
```

buat prediksi

```python
pred = lr.predict(X_test)
pred
```

cetak akurasi dari data testingnya

```python
print("Akurasi dari data Testing :",lr.score(X_test, y_test))
```

periksa errornya pake MSE

```python
from sklearn.metrics import mean_squared_error

print("Mean Squared Error dari Data Testing :",mean_squared_error(y_test,pred))
```
